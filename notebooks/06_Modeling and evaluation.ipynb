```python
from sklearn.metrics import recall_score, precision_score

# Calculate recall with different averaging methods
recall_macro = recall_score(y_true=y_test, y_pred=y_test_predicted, average='macro')
print("Recall (macro):", recall_macro)
# recall_weighted = recall_score(y_true=y_test, y_pred=y_test_predicted, average='weighted')

# Calculate precision with different averaging methods
precision_macro = precision_score(y_true=y_test, y_pred=y_test_predicted, average='macro')
print("Precision (macro):", precision_macro)        
# precision_weighted = precision_score(y_true=y_test, y_pred=y_test_predicted, average='weighted')
```

```python
# Get scores for each class individually
recall_per_class = recall_score(y_true=y_test, y_pred=y_test_predicted, average=None)
precision_per_class = precision_score(y_true=y_test, y_pred=y_test_predicted, average=None)

classes = np.unique(y_test)
for i, cls in enumerate(classes):
    print(f"{cls}:")
    print(f"  Recall:    {recall_per_class[i]:.4f}")
    print(f"  Precision: {precision_per_class[i]:.4f}")
```

### MODEL 4 : Multi-Layer Perceptron / MLP (non-linear)

```python
from sklearn.neural_network import MLPClassifier
```

```python
X_train.shape
```

```python
# mlp_model 1, rule of thumb: Start with (n_featurs + n_classes) / 2 = (50+3)/2 = 27
mlp_model = MLPClassifier(hidden_layer_sizes=(27,15),max_iter=1000, random_state=0)
mlp_model.fit(X=X_train, y=y_train)
y_test_predicted_mlp = mlp_model.predict(X_test)
report_mlp = classification_report(y_true=y_test, y_pred=y_test_predicted_mlp)
print(report_mlp)
```

```python
# mlp_model 2, More complex, 3 layers
mlp_model2 = MLPClassifier(hidden_layer_sizes=(100,50,25),max_iter=1000, random_state=0)
mlp_model2.fit(X=X_train, y=y_train)
y_test_predicted_mlp2 = mlp_model2.predict(X_test)
report_mlp2 = classification_report(y_true=y_test, y_pred=y_test_predicted_mlp2)
print(report_mlp2)
```

### Model 5: XGBoost

```python
import xgboost as xgb
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder

# Encode target labels for XGBoost
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

print("Label mapping:")
for i, label in enumerate(label_encoder.classes_):
    print(f"  {label} -> {i}")

# XGBoost Model
xgb_model = xgb.XGBClassifier(
    n_estimators=200,
    max_depth=6,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=0,
    eval_metric='mlogloss'
)

# Train with encoded labels
xgb_model.fit(X_train, y_train_encoded)

# Predict (returns numeric labels)
y_test_predicted_xgb_encoded = xgb_model.predict(X_test)

# Convert predictions back to original labels
y_test_predicted_xgb = label_encoder.inverse_transform(y_test_predicted_xgb_encoded)

# Evaluate with original string labels
report_xgb = classification_report(y_true=y_test, y_pred=y_test_predicted_xgb)
print("XGBoost Results:")
print(report_xgb)
```
