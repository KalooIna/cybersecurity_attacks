```python
# Opt1 : Drop Device_family and Packet Length_Bin
df = df.drop(columns=['Device_family', 'Packet Length_Bin'])
```

```python
# Take only top 3 of ContinentName_Source, ContinentName_Destination and put others as 'Other'
top_continents_source = df['ContinentName_Source'].value_counts().nlargest(3).index
df['ContinentName_Source'] = df['ContinentName_Source'].apply(lambda x: x if x in top_continents_source else 'Other')
```

```python
top_continents_destination = df['ContinentName_Destination'].value_counts().nlargest(3).index
df['ContinentName_Destination'] = df['ContinentName_Destination'].apply(lambda x: x if x in top_continents_destination else 'Other')
```

```python
top_continents_proxy = df['ContinentName_Proxy'].value_counts().nlargest(3).index
df['ContinentName_Proxy'] = df['ContinentName_Proxy'].apply(lambda x: x if x in top_continents_proxy else 'Other')
```

```python
# 'Browser_family', ==> non ordinal ==> 9 unique values, top 5 occupies 81.51%, rest 18.49% (mobile browser family) ==> take top 5 and rest as 'Other'
top_browser_family = df['Browser_family'].value_counts().nlargest(5).index
df['Browser_family'] = df['Browser_family'].apply(lambda x: x if x in top_browser_family else 'Other')
```
