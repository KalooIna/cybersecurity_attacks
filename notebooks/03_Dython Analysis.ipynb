```python
# Load df processed after first Feature Engineering & EDA
df = pd.read_csv('../data/processed_data/cybersecurity_attacks_firstFE&EDA.csv')
```

```python
# annot = annotation to show numbers in the plot
# .2f to show 2 decimal places
# Dython use Pearson for numerical and Cramer's V for categorical by default
# This will take time to run.
# in plot, SV means Single Value (only one unique value) so Dython cannot compute association
# Most heat seen between new features created from same columns, ie. Device_type - Device_family, Proxy_Class - ContinentName_Proxy, etc
# Meanwhile between other features seem low, like 0.01, 0.0 (the highest) and lots 0.000x - this prove that need combinations of features to be able to predict the target
corr_matrix = associations(df, figsize=(15, 15), cmap="coolwarm", annot=True, fmt=".2f")
plt.show()
```
